```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import wasserstein_distance
from careamics.noise_model import NoiseModelTrainer
```

```python
signal = ...
observation = ...
```

```python
trainer = NoiseModelTrainer(n_gaussian=3, n_coeff=3, min_sigma=100.0)
trainer.train_from_pairs(
    signal=signal,
    observation=observation,
    n_epochs=2000,
    learning_rate=0.1,
)
```

```python
sampled = trainer.sample_observation(signal)
```

```python
real_noise = (observation - signal).ravel()
synth_noise = (sampled - signal).ravel()

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].hist(real_noise, bins=100, alpha=0.7, label="Real")
axes[0].hist(synth_noise, bins=100, alpha=0.7, label="Sampled")
axes[0].legend()
axes[0].set_title("Noise Distribution")

axes[1].imshow(observation[0] if observation.ndim > 2 else observation)
axes[1].set_title("Original Noisy")

axes[2].imshow(sampled[0] if sampled.ndim > 2 else sampled)
axes[2].set_title("Sampled Noisy")

plt.tight_layout()
plt.show()
```

```python
scale = signal.max() - signal.min()
distance = wasserstein_distance(real_noise / scale, synth_noise / scale)
print(f"Wasserstein distance: {distance:.4f}")
```
